# Data Preprocessing

```python
# Drop columns irrelevant to training
cols_to_drop = ['PassengerId', 'Name', 'Cabin']

df = df.drop(cols_to_drop, axis=1)
```

```python
# Convert categorical columns using one-hot encoding 

categorical_columns = ['HomePlanet', 'Destination']

for col in categorical_columns:
    df = pd.concat([df, pd.get_dummies(df[col], dtype=int)], axis=1)


# Then, drop columns that have been processed
df = df.drop(categorical_columns, axis=1)
```

```python
# Fill NaN values
df.isna().sum()
```

```python
imputer = SimpleImputer()
cols = df.columns
df = pd.DataFrame(imputer.fit_transform(df), columns=cols)
```

```python
# Converting boolean columns to integers
bool_cols = ['CryoSleep', 'VIP', 'Transported']

for col in bool_cols:
    df[col] = df[col].astype(int)
```

```python
# View cleaned train data
df.head()
```

```python
df.to_csv('./data/train-cleaned.csv', index=None)
```
